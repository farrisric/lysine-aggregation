```python
import mdtraj as md
import numpy as np

work_dir = '/media/riccardo/UNICA/Lysine/diamanda_stuff/20/'

traj = md.load(work_dir+'rstrip.nc', top=work_dir+'strip.prmtop')
topology = traj.topology
```

```python
c_atoms = np.array([atom.index for atom in topology.atoms if atom.element.symbol == 'C'])
c_atoms = c_atoms.reshape(traj.n_residues,6)
c_atoms[:,2]
```

```python
from mdtraj import compute_neighbors

alfa_carbons = c_atoms[:,2]

for alfa_carbon in alfa_carbons:
    ciao = compute_neighbors(traj,0.5,[alfa_carbon],alfa_carbons)
```

```python
for frame, f in enumerate(ciao):
    if len(f) > 0 :
        print(frame, alfa_carbon,f)
```
