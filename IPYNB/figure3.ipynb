```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cm = 1/2.54  # centimeters in inches
plt.rcParams['figure.dpi'] = 200
font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 8}
```

```python
gromacs_x = np.loadtxt('../data/diff_coeff_gromacs.txt')
amber_diff = np.loadtxt('../data/diff_coeff_amber.txt')
```

```python
# Data
x1 = np.array([10, 50, 100, 200, 500])/1000
y1 = np.array([5.96, 5.77, 5.60, 5.29, 4.71])

x2 = np.array([0.12, 0.25, 0.38, 0.51, 0.64, 0.77, 0.90, 1.04])
y2 = np.array([1.4840, 1.2112, 1.2027, 0.9063, 0.9701, 0.9383, 0.8659, 0.7487])*10

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(8.54*cm,6.54*cm))

# Plot the data
ax.plot(x1, y1, '-o', label='Exp.')
ax.plot(x2, y2,'-o', label='Amber old')
ax.plot(x2, amber_diff*10,'-o', label='Amber new')
ax.errorbar(x2, gromacs_x[:,1], gromacs_x[:,2], marker='o', label='Gromas new')
#ax.errorbar(x2,gromacs_x[:,1], gromacs_x[:,2])
ax.legend(bbox_to_anchor=(1.275, 0.95), loc='upper left', borderaxespad=0.05, ncol=1)

ax.set_ylabel('Dif. Coef. [10$^{-10}$ m$^{2}$/s]')
ax.set_xlabel('Lysine Concentration [M]')
#plt.tight_layout()
```
