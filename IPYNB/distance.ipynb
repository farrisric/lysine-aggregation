```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from glob import iglob
```

```python
work_dir = 'D:/work/lysine/Lysine/diamanda_stuff/'

for distance_out in iglob(f'{work_dir}/*/dist[1-8]0.dat'):
    X = np.loadtxt(distance_out)
    data = X[:,1:].flatten()
    ax = sns.kdeplot(data)
    lines = ax.get_lines()
    x = lines[0].get_xdata()
    y = lines[0].get_ydata()
    name = distance_out.split('/')[-1].split('\\')[-1].split('.')[0]
    np.savetxt(f'../data/{name}.txt', (x,y))
    plt.show()
```

```python
cm = 1/2.54  # centimeters in inches
plt.rcParams['figure.dpi'] = 200
font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 8}
```

```python
fig, axes = plt.subplots(figsize=(8.5*cm,8.5*cm))

colors = list(reversed(['#d73027','#f46d43','#fdae61','#fee090','#e0f3f8','#abd9e9','#74add1','#4575b4'])) 
concentrations = [0.12,0.25,0.38,0.51,0.64,0.77,0.91,1.04,]

for i,f in enumerate(iglob(f'../data/dist[1-8]0.txt')):
    data = np.loadtxt(f)
    axes.plot(data[0], data[1], color=colors[i], linewidth=1.5, label=str(concentrations[i])+' M')
    
axes.legend()
axes.set_xlim(3,8)
axes.set_ylim(0,1.4)
axes.set_ylabel('Density')
axes.set_xlabel('r$_{C1-C6}$ [$\AA$]')
```
