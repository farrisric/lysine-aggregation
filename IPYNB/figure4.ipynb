```python
import numpy as np
import matplotlib.pyplot as plt
from glob import iglob
```

```python
cm = 1/2.54  # centimeters in inches
plt.rcParams['figure.dpi'] = 200
font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 8}
```

```python
fig, axis = plt.subplots(3,2, figsize = (8.25*cm, 12.75*cm), sharex=True, sharey=True)
colors = list(reversed(['#d73027','#f46d43','#fdae61','#fee090','#e0f3f8','#abd9e9','#74add1','#4575b4'])) 
angles = ['zero', 'alfa', 'beta', 'gamma', 'delta']
titles = ['α-1', 'α', 'β', 'γ', 'δ']
concentrations = [0.12,0.25,0.38,0.51,0.64,0.77,0.91,1.04,]

axis = axis.flat
g=0
for angle, axes in zip(angles, axis):
    axes.annotate(titles[g], (30,0.025))
    g+=1
    for i,f in enumerate(iglob(f'{angle}*txt')):
        data = np.loadtxt(f)
        axes.plot(data[0], data[1], color=colors[i], linewidth=1.5, label=str(concentrations[i])+' M')
        axes.set_xticks(np.rad2deg(axes.get_xticks()))

axis[-2].legend(bbox_to_anchor=(1.275, 0.95), loc='upper left', borderaxespad=0.05, ncol=1)
axis[-1].set_visible(False)

axis[3].xaxis.set_visible(True)
axis[3].set_xticks([0,90,180,270,360])

axis[0].set_ylabel('Density')
axis[2].set_ylabel('Density')
axis[4].set_ylabel('Density')

axis[3].set_xlabel('Degrees')
axis[4].set_xlabel('Degrees')


plt.ylim(0,0.03)
plt.xlim(0,360)
_ = plt.xticks([0,90,180,270,360])
```
